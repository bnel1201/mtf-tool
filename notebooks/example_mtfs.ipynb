```python
import numpy as np
import matplotlib.pyplot as plt

from xri_mtf import mtf

ideal = mtf.from_csv('ideal_edge_sim.csv').rename(columns={'MTF': 'ideal'})
blur = mtf.from_csv('blurred_edge_sim.csv').rename(columns={'MTF': 'blurred'})
mtfs  = ideal.set_index('frequency [1/mm]').join(blur.set_index('frequency [1/mm]'))
mtfs.head()
```

```python
f, ax = plt.subplots(dpi=150)
mtfs.plot(ax=ax)
ax.set_xlim([0, 100])
```
